```python
%%writefile app.py
import pandas as pd
import matplotlib.pyplot as plt
import streamlit as st

st.title("Trading Dashboard Mobile")

n_days = 50

data = {
    'Date': pd.date_range(start='2026-01-01', periods=n_days, freq='D'),
    'Close': [10+i%10 for i in range(n_days)]
}

df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

df['SMA_5'] = df['Close'].rolling(5).mean()
df['SMA_10'] = df['Close'].rolling(10).mean()

fig, ax = plt.subplots()
ax.plot(df.index, df['Close'], label='Close')
ax.plot(df.index, df['SMA_5'], label='SMA 5')
ax.plot(df.index, df['SMA_10'], label='SMA 10')
ax.legend()

st.pyplot(fig)
st.dataframe(df)
```
